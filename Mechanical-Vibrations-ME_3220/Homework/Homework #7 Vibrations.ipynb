# Problem 3.11

Calculate the response of:

$$3\ddot{x}+12\dot{x}+12x(t)=3\delta(t)$$

for zero initial conditions.

## Solution

We will assume this system has the following components:

$m = 3$

$c = 12$

$k = 12$

Where this system properties are:

$\omega_n = \sqrt{\frac{12}{3}}=2$

$\zeta = \frac{12}{2\sqrt{12\cdot3}}=1$

That means this system is critcally damped: and the solution will have the general form:

$$x(t) = (a_1+ta_2)e^{-\omega_n t}$$

Where due to the impulse:

$x_o = 0$

$v_o = \frac{\hat{F}}{m} = 1$

And our constants become:

$a_1 = x_o = 0$

$a_2 = v_o + \omega_n x_o = 1$

This means our response function is:

$$x(t) = te^{-2t}$$

We can model this system as being:

$$m\ddot{x} + k x = F(t)$$

We will also say that the homogenous solution has solution:

$$h(t) = \frac{1}{m\omega_n}sin(\omega_n t)$$

Using convolution, we will say that:

$$x(t) = \int_o^{t_o} F_o\left(1-\frac{\tau}{t_o}\right)\frac{1}{m\omega_n}sin(\omega_n (t-\tau))d\tau$$

We can isolate anything that isn't a function of $\tau$ out to the front and split the integral:

$$x(t) = \frac{F_o}{m\omega_n}\int_o^{t_o}sin(\omega_n(t-\tau))d\tau - \frac{F_o}{mt_o\omega_n}\int_o^{t_o}\tau sin(\omega_n(t-\tau))d\tau$$

We will simplify this first to:

$$\frac{F_o}{m\omega_n^2}cos(\omega_n(t-\tau))\left|_{\tau=0}^{\tau=t_o}-\left(\frac{\tau F_o}{mt_o\omega_n}cos(\omega_n(t-\tau))\right|_{\tau=0}^{\tau=t_o}-\frac{F_o}{mt_o\omega_n}\int_0^{t_o}cos(\omega_n(t-\tau))d\tau\right)$$

$$\frac{F_o}{m\omega_n^2}(cos(\omega_n(t-t_o))-cos(\omega_n(t)))-\left(\frac{F_ot_o}{mt_o\omega_n^2}cos(\omega_n(t-t_o))+\frac{F_o}{mt_o\omega_n^2}\left(sin(\omega_n(t-t_o))-sin(\omega t)\right)
  \right)$$

We have terms that cancel out leaving:

Where $m\omega_n^2 = k$ so:

# Problem 3.33

For a Fourier series, we know that any function can be modelled as:    

$$F(t) = \frac{a_o}{2}+\sum_{n=1}^{\infty}\left(a_ncos(n\omega_Tt)+b_nsin(n\omega_Tt)\right)$$

where the constants are defined by:

$a_o = \frac{2}{T}\int_0^Tf(t)dt$

$a_n = \frac{2}{T}\int_0^Tf(t)cos(n\omega_Tt)dt$

$b_n = \frac{2}{T}\int_0^Tf(t)sin(n\omega_Tt)dt$

In our case:

$T = 2\pi$

$\omega_T = \frac{2\pi}{2\pi}=1$

$f(t) = \frac{1}{2\pi}t$

Plugging this in gives:    

$$a_0 = \frac{1}{2}$$

$$a_n = \frac{2}{(2\pi)^2}\left(\frac{2\pi}{n}sin(2\pi t)+\frac{1}{n^2}(cos(2\pi n)-cos(0))\right)$$

for integer values of $n$ this simplifies to:     

$$a_n = 0$$

and then:

$$b_n = \frac{2}{(2\pi)^2}\left(\frac{-2\pi}{n}cos(2\pi n)+\frac{1}{n^2}sin(2\pi n)\right)$$

For constant values of $n$:     

$$b_n = \frac{-2}{2\pi n}$$

This gives a full fourier series of:

$$F(t) = \frac{a_o}{2}+\sum_{n=1}^{\infty}\left(a_ncos(nt)+b_nsin(nt)\right)$$

# Problem 4

Solve the system:

$$\ddot{x}+2\dot{x}+2x=\delta(t-2\pi)+cos(t+2\pi)+sin(t+2\pi)$$

for initial conditions $x(0)=0.1, v(0)=0.1$.


We can break this problem into 4 seperate problems. First we will solve the initial condition problem:    

$$\ddot{x} + 2\dot{x} + 2x = 0,\qquad x(0) = 0.1,\dot{x} = 0.1$$

Which has a solution of:

$$x(t) = e^{-t}(0.2sin(t)+0.1cos(t))$$

because $\zeta = \frac{1}{\sqrt{2}}$, $\omega_n = \sqrt{2}$, and $\omega_d = 1$.

The next problem we will look at is:      

$$\ddot{x} +2\dot{x}+2x=\delta(t-2\pi)$$

this has solution

$$x(t)={e^{(-t+2\pi)}}sin(t-2\pi)$$


We will then look at

$$\ddot{x} +2\dot{x}+2x=cos(t+2\pi)$$

comparing this solution to our harmonic response, we know that

$$x(t)=A_scos(t+2\pi)+B_ssin(t+2\pi)$$

Where:

$A_s = \frac{(2-1)}{(2-1)^2+(2)^2}=\frac{1}{5}$

$B_s = \frac{2}{(2-1)^2+(2)^2}=\frac{1}{5}$

We will lastly say our final equation is

$$\ddot{x} +2\dot{x}+2x=cos(t-5\pi/2)$$

since $sin(t) = cos(t-\pi/2)$

comparing this solution to our harmonic response, we know that

$$x(t)=A_scos(t+3\pi/2)+B_ssin(t+3\pi/2)$$

Where:

$A_s = \frac{(2-1)}{(2-1)^2+(2)^2}=\frac{1}{5}$

$B_s = \frac{2}{(2-1)^2+(2)^2}=\frac{1}{5}$

This means our total solution is  

$$x(t) = e^{-t}(0.2sin(t)+0.1cos(t)) + {e^{(-t+2\pi)}}sin(t-2\pi) + \frac{1}{5}\left(cos(t+2\pi)+sin(t+2\pi) + cos(t+3\pi/2)+sin(t+3\pi/2)\right)$$
