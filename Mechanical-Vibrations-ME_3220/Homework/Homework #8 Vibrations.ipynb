# Problem 1

Consider the system depicted below. For $c_1=c_2=c_3=0$, derive the equation of motion and calculate the mass and stiffness matrices.

# Solution

We will calculate the free body diagram for each mass

FBD$_1$

$$\sum F=m_1\ddot{x}_1$$

$$-k_1x_1-k_2(x_1-x_2)=m_1\ddot{x}_1$$

$$m_1\dot{x}_1+k_1x_1+k_2x_1-k_2x_2=0$$

FBD$_2$

$$\sum = F=m_2\ddot{x}_2$$

$$-k_2(x_2-x_1)-k_3x_2=m_2\dot{x}_2$$

$$m_2\ddot{x}_2+k_2x_2-k_2x_1+k_3x_2=0$$

This gives us a system of equations

$$\begin{cases} 
      m_1\ddot{x}_1+(k_1+k_2)x_1+(-k_2)x_2=0\\
      m_2\ddot{x}_2+(-k_2)x_1+(k_2+k_3)x_2=0
  \end{cases}
  $$

This can be rewritten in matrix form as

$$\begin{bmatrix} m_1 & 0 \\ 0 & m_2\end{bmatrix} \begin{bmatrix}\ddot{x}_1\\\ddot{x}_2\end{bmatrix} + \begin{bmatrix}k_1+k_2 & -k_2 \\ -k_2 & k_2+k_3\end{bmatrix} \begin{bmatrix}x_1\\x_2\end{bmatrix} = \vec{0}$$

# Problem 2

Calculate the characteristic equation for Problem 1 for the case

$$m_1=9 kg\qquad m_2 = 1 kg\qquad k_1=24N/m\qquad k_2 = 3N/m \qquad k_3 = 3 N/m$$

and solve for the system's natural frequencies.

# Solution

We will assume our solution has the form

$$\vec{x}=\vec{v}e^{-i\omega t}$$

Which transforms our equation into

$$-\omega^2M\vec{v}e^{-i\omega t} + K\vec{v}e^{-i\omega t} = 0$$

Which simplifies to

$$(-\omega^2M+K)\vec{v}=0$$

For this problem, our matricies are

$$M = \begin{bmatrix}9 && 0\\0&&1\end{bmatrix}$$

$$K = \begin{bmatrix}27 && -3\\-3&&6\end{bmatrix}$$

To solve this equation, the determinate must be 0

$$det\left(-\omega^2M+K\right)=0$$

This gives us the 4th degree polynomial

$$9\omega^4-81\omega^2+153=0$$

We will make a change of varibles here, saying $\omega^2=\lambda$ so our equation becomes a second degree polynomial


$$9\lambda^2-81\lambda+153=0$$

Which has roots

$$\lambda_1 = \frac{9-\sqrt{13}}{2}\approx2.697$$

$$\lambda_2 = \frac{9+\sqrt{13}}{2}\approx6.303$$

This means that natural frequencies are

$$\omega_1 = \sqrt{\frac{9-\sqrt{13}}{2}}\approx1.64$$

$$\omega_2 = \sqrt{\frac{9+\sqrt{13}}{2}}\approx2.51$$

# Problem 3

Calculate the vectors $\vec{u}_1$ and $\vec{u}_2$ for Problem 2.

# Solution

To find our eigenvectors, we will plug our frequencies back into the equation

$$(-\lambda M+K)\vec{u}=0$$

We will handle this seperatly for our two frequencies. Dealing with our first one ($\lambda_1$)

$$(-\lambda_1M+K)=\begin{bmatrix}\frac{9}{2}(-3+\sqrt{13})&-3\\ -3 & \frac{1}{2}(3+\sqrt{13})\end{bmatrix}$$

We know these are always linearly dependent so we will just use the first relation to say that

$$2.725u_{1,1}-3u_{1,2}=0$$

$$u_{1,1}=1.1u_{1,2}$$

This means our vector is

$$\vec{u}_1=\begin{bmatrix}u_{1,1}\\u_{1,2}\end{bmatrix}=\begin{bmatrix}1.1u_{1,2}\\u_{1,2}\end{bmatrix}=\alpha\begin{bmatrix}1.1\\1\end{bmatrix}$$

Our next natural frequency leads to

$$(-\lambda_2M+K)=\begin{bmatrix}\frac{-9}{2}(3+\sqrt{13})&-3\\ -3 & \frac{1}{2}(3-\sqrt{13})\end{bmatrix}$$

Again, these are linearly dependent so we will take the first row to say

$$-29.725v_{2,1}-3u_{2,2}=0$$

$$u_{2,1} = -0.1u_{2,2}$$

Which gives our vector

$$\vec{u}_2=\begin{bmatrix}u_{2,1}\\u_{2,2}\end{bmatrix}=\begin{bmatrix}-0.1u_{2,2}\\u_{2,2}\end{bmatrix}=\beta\begin{bmatrix}-0.1\\1\end{bmatrix}$$

# Problem 4

For the initial conditions $\vec{x}(0)=[1\quad 0]^T$ and $\vec{\dot{x}}(0)=[0\quad 0]^T$ calcualte the free response of the system from Problem 2. Plot the response $x_1(t)$ and $x_2(t)$.

# Solution

We now know our system has the solution

$$x(t) = A_1\vec{u}_1cos(\omega_1t)+B_1\vec{u}_1sin(\omega_1t)+A_2\vec{u}_2cos(\omega_2t)+B_2\vec{u}_2sin(\omega_2t)$$

Plugging in the vectors and breaking appart the matrix gives

$$x_1(t)=1.1A_1cos(\omega_1t)+1.1B_1sin(\omega_1t)-0.1A_2cos(\omega_2t)-0.1\vec{u}_2sin(\omega_2t)$$

$$x_2(t)=A_1cos(\omega_1t)+B_1sin(\omega_1t)+A_2cos(\omega_2t)+B_2sin(\omega_2t)$$

We can plug in our first set of initial conditions here

$$1=1.1A_1-0.1A_2$$

$$0=A_1+A_2$$

Solving this give

$$A_1 = \frac{5}{6}$$

$$A_2 = \frac{-5}{6}$$

We will now take the deriviatives of $\vec{x}$ and plug in the initial conditions

$$0=1.1\omega_1B_1-0.1\omega_2B_2$$

$$0=\omega_1B_1+\omega_2B_2$$

Which has trivial solutions

$$B_1 = 0$$

$$B_2 = 0$$

Which gives us an over all equation of

$$\vec{x}(t)=\frac{5}{6}\vec{u}_1cos(\omega_1t)-\frac{5}{6}\vec{u}_2cos(\omega_2t)$$

$$\sum F_{m_2}=<m_2\ddot{x_3},m_2\ddot{x_4}>$$

On $m_1$ we have 4 forces

$$F_{Left},F_{Right},F_{Top},F_{Bottom}$$

Where for an offest mass

The equations of motion for this system are

$$\sum F_{x_1}=m\ddot{x}_1$$

$$\sum F_{x_2}=m\ddot{x}_2$$

$$\sum F_{x_3}=m\ddot{x}_3$$

$$\sum F_{x_4}=m\ddot{x}_4$$

Which are explicity given by

$$-4kx_1-k(x_1-x_3)=m\ddot{x}_1$$

$$-k(x_3-x_1)-4x_3=m\ddot{x}_3$$

$$-2kx_4-4kx_4=m\ddot{x}_4$$

$$m\ddot{x}_1+5kx_1-kx_3=0$$

$$m\ddot{x}_2+4kx_2=0$$

$$m\ddot{x}_3+5kx_3-kx_1=0$$

$$m\ddot{x}_4+6kx_4=0$$

$$M\vec{\ddot{x}}+K\vec{x}=0$$
