## Problem 1:

Compute the initial conditions such that the response of 

$$m\ddot{x}+kx=F_ocos(\omega t)$$

ocsillates at only one frequency ($\omega$).



### Solution

This general harmonic excitation equation has a solution with the form:

$$x(t)=A_1cos(\omega_nt)+A_2sin(\omega_nt)+\frac{f_o}{\omega_n^2-\omega^2}cos(\omega t)$$

Where:

$f_o = \frac{F_o}{m}$

$\omega_n = \sqrt{\frac{k}{m}}$

$A_1 = x_o-\frac{f_o}{\omega_n^2-\omega^2}$

$A_2 = \frac{v_o}{\omega_n}$

If we want our solution to only oscillate at $\omega$ then we must ensure that our trigonometric functions without frequencies of $\omega$ are cancled out. We can do this by having $A_1=0$ and $A_2=0$.

For $A_2=0$ we can simply have an initial velocity of $v_o=0$. To have $A_1=0$ we must have our inital displacement $x_0=\frac{f_o}{\omega_n^2-\omega^2}$.

### Solution

Since the springs are in parallel we can combind there spring constants ($K=k_1+k_2$). This system has the equation:

$$\ddot{x}+\frac{c}{m}\dot{x}+\frac{K}{m}x=f_ocos(\omega t)$$

We can transform this equation to the form:

$$\ddot{x}+2\zeta \omega_n \dot{x}+\omega_n^2x=f_ocos(\omega t)$$

Where:

$\zeta = \frac{c}{2\sqrt{Km}}$

$\omega_n=\sqrt{\frac{K}{m}}$

$f_o=\frac{F_o}{m}$

This equation has the general solution form of:

$$x(t)=e^{-\zeta \omega_n t}\left(A_1cos(\omega_d t)+A_2sin(\omega_d t)\right)+A_scos(\omega t)+B_ssin(\omega t)$$

Where:

$A_1 = x_0-\frac{(\omega_n^2-\omega^2)f_o}{(\omega_n^2-\omega^2)^2+(2\zeta\omega_n\omega)^2}$

$A_2=\frac{\zeta\omega_n}{\omega_d}\left(x_0-\frac{(\omega_n^2-\omega^2)f_o}{(\omega_n^2-\omega^2)^2+(2\zeta\omega_n\omega)^2}\right)-\frac{2\zeta\omega_n\omega^2f_o}{\omega_d\left((\omega_n^2-\omega^2)^2+(2\zeta\omega_n\omega)^2\right)}+\frac{v_o}{\omega_d}$

$A_s = \frac{(\omega_n^2-\omega^2)f_o}{(\omega_n^2-\omega^2)^2+(2\zeta\omega_n\omega)^2}$

$B_s = \frac{2\zeta\omega_n\omega f_o}{(\omega_n^2-\omega^2)^2+(2\zeta\omega_n\omega)^2}$

```python
import numpy as np
from pint import UnitRegistry
ureg = UnitRegistry()
Q_ = ureg.Quantity
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
k_1 = Q_(100,"N/m")
k_2 = Q_(500,"N/m")
K = 1/(1/k_1 + 1/k_2)
c = Q_(20,"kg/s")
m = Q_(89,"kg")
F_o = Q_(25,"N")
f_o = F_o/m
omega = Q_(3,"rad/s")
x_o = Q_(0, "m")
v_o = Q_(0, "m/s")
omega_n = np.sqrt(K/m).to("rad/s")


zeta = c/(2*np.sqrt(K*m))
omega_d = (omega_n*np.sqrt(1-zeta**2)).to("rad/s")

A_1 = x_o - ((omega_n**2-omega**2)*f_o)/((omega_n**2-omega**2)**2+(2*zeta*omega_n*omega)**2)
A_2 = (zeta*omega_n/omega_d)*(A_1)-(2*zeta*omega_n*omega**2*f_o)/(omega_d*((omega_n**2-omega**2)**2+(2*zeta*omega_n*omega)**2))+v_o/omega_d
A_s = ((omega_n**2-omega**2)*f_o)/((omega_n**2-omega**2)**2+(2*zeta*omega_n*omega)**2)
B_s = (2*zeta*omega_n*omega*f_o)/((omega_n**2-omega**2)**2+(2*zeta*omega_n*omega)**2)

def x(t):
  return np.exp(-zeta*omega_n*t)*(A_1*np.cos(omega_d*t)+A_2*np.sin(omega_d*t)) + A_s*np.cos(omega*t) + B_s*np.sin(omega*t)

t = np.linspace(0,50,5000)*ureg.seconds
plt.plot(t,x(t))
plt.grid(True)
plt.show()
```

Since the system is initially at rest, we will assume $x_0=0$ and $v_0=0$.
