## Problem 1:

Consider a spring-mass-damper system, like the one in Figure 1.9 (depicted below), with the following values: $m=10$ kg, $c=3$ N/s, and $k=1000$ N/m.

(b) Compute the solution if the system is given initial conditions $x_0=0.01$ m and $v_0$ = 0.

### Solution:

We can use the damping ration $\zeta$ to determine the damping of our system:

$$\zeta = \frac{c}{2\sqrt{km}}$$

```python
import numpy as np
import matplotlib.pyplot as plt 
m = 10
c = 3
k = 1000
zeta = c/(2*np.sqrt(k*m))
print(zeta)
```

Here we have a damping of 0.015 which means our system is underdamped since $0<\zeta <1$.

For a system that is underdamped, the general solution is:

$$x(t)=e^{-\zeta \omega_nt}\left[A_1sin(\omega_dt)+A_2cos(\omega_dt)\right]$$

Where:

$\qquad A_1 = \frac{v_0+\zeta \omega_n x_0}{\omega_d}$

$\qquad A_2 = x_0$

```python
omega_n = np.sqrt(k/m)
omega_d = omega_n*np.sqrt(1-zeta**2)
print("Natural frequency =", omega_n)
print("Damped natural frequency =", omega_d)

x_0 = 0.01
v_0 = 0

A_1 = (v_0 + zeta*omega_n*x_0)/omega_d
A_2 = x_0

print("A_1 =", A_1)
print("A_2 =", A_2)
t = np.linspace(0,10*2*np.pi/omega_d,200)
def x(t):
  return np.exp(-zeta*omega_n*t)*(A_1*np.sin(omega_d*t)+A_2*np.cos(omega_d*t))
plt.plot(t,x(t))
plt.grid(True)
```

The general solution would be:

$$x(t) = e^{-0.15t}[0.00015sin(10t)+0.01cos(10t)]$$

### Solution

The damping ratio is described by:

$$\zeta = \frac{c}{2\sqrt{km}}$$

The damped natural frequency is described by: 

$$\omega_d = \omega_n \sqrt{1-\zeta^2}$$

The undamped natural frequency is described by:

$$\omega_n = \sqrt{\frac{k}{m}}$$

```python
m = 100
c = 300
k = 3000

zeta = c/(2*np.sqrt(k*m))
omega_n = np.sqrt(k/m)
omega_d = omega_n*np.sqrt(1-zeta**2)

print("Undamped natural frequency =",omega_n.round(2))
print("Damping ratio =", zeta.round(2))
print("Damped natural frequency =",omega_d.round(2))
```
