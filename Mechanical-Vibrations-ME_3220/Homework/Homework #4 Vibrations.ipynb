# Problem 1

A control pedal of an aircraft can be modeled as the single-degree-of-freedom system pictured below. Consider the lever as a massless shaft and the pedal as a lumped mass at the end of the shaft. Use the energy method to determine the equation of motion in $\theta$ and calculate the natural frequency of the system. Assume the spring to be unstretched at $\theta=0$ and gravity points down.

Screen Shot 2021-09-27 at 7.20.25 PM.png

## Solution

For the energy method with no work, we know:

$$\sum U+\sum T=Constant$$

and differentiating that give:

$$\frac{dU}{dt}+\frac{dT}{dt}=0$$

For our system, the potential energys are:

$$U_{spring}=\frac{1}{2}kx^2$$

$$U_{gravity} = mgh$$

Where the following trigometric equations apply:

$$\frac{x}{l_1}=sin(\theta)$$

$$\frac{l_2-h}{l_2}=cos(\theta)$$

Plugging these in to our potential energy equations gives:

$$U_{spring}=\frac{1}{2}k(l_1 sin(\theta))^2$$

$$U_{gravity} = mg(l_2-l_2 cos(\theta))$$

Our kinetic energy equation is:

$$T=\frac{1}{2}J\dot{\theta}^2$$

Where we can say:

$$\dot{x}=l_2\dot{\theta}$$

So we will rearrange the kinetic energy to be:

$$T=\frac{1}{2}m(l_2\dot{\theta})^2$$

This gives a total energy equation of:

$$\frac{1}{2}k(l_1 sin(\theta))^2+mg(l_2-l_2 cos(\theta))+\frac{1}{2}m(l_2\dot{\theta})^2=constant$$

Now we will differentiate this equation to get:

$$k(l_1)^2 sin(\theta)cos(\theta)\dot{\theta}+mgl_2sin(\theta)\dot{\theta}+m(l_2)^2\ddot{\theta}\dot{\theta}=0$$

Linearizing this equation gives and dividing by $\dot{\theta}$:

$$k(l_1)^2\theta+mgl_2\theta+m(l_2)^2\ddot{\theta}=0$$

And we can rewrite this into the proper form for an equation of motion:

$$ml_2^2\ddot{\theta}+(mgl_2+kl_1^2)\theta=0$$

The natural frequency would be:

$$\omega_n=\sqrt{\frac{mgl_2+kl_1^2}{ml_2^2}}$$

$$\frac{x}{l}=sin(\theta)$$

$$\frac{h}{\frac{l}{2}}=cos(\theta)$$

We can now rearrange our potential energy equations as:

$$U_{spring} = \frac{1}{2}kl^2sin^2(\theta)$$

$$U_{gravity} = mg\frac{l}{2}cos(\theta)$$

And we know the kinetic energy would be:

$$T=\frac{1}{2}\frac{ml^2}{3}\dot{\theta}^2$$

## Solution 

Using Newtons second law, we can say

$$\sum F = mgsin(\theta)-kx-\mu mgcos(\theta)sign(\dot x)=m\ddot{x}$$

This means we have the equation of motion:

$$m\ddot{x}+kx=mgsin(\theta)-\mu mgcos(\theta)sign(\dot{x})$$

To solve this equation, we will write it in the form of a first order state space model where it can then be solved in MATLAB. First we will change the form of the equation to:

$$\ddot{x}=gsin(\theta)-\frac{k}{m}x-\mu gcos(\theta)sign(\dot{x})$$

Then we can say our solution $x=x_1$ and therefore:

$$\dot{x_1} = x_2$$

$$\dot{x_2} = gsin(\theta)-\frac{k}{m}x_1-\mu gcos(\theta)sign(x_2)$$
