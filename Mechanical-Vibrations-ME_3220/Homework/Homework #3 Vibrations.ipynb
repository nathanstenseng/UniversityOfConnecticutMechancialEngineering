## Problem 1

Use the engery method to calculate the equation of motion and natural frequency of an airplane's steering mechanism for the nose wheel of its landing gear. The machanism is modelded as the single-degree-of-freedom system illistrated below:

Screen Shot 2021-09-21 at 9.31.33 PM.png

The steering wheel and tire assembly are modeled as being fixed at ground for this calculation. The steering rod gear system is modeled as a linear spring-mass system ($m,k_2$) oscillating in the x direction. The shaft-gear mechanism is modeled as the disk of inertia $J$ and torional stiffness $k_1$. The gear $J$ turns through the angle $\theta$ such that the disk does not slip on the mass. Obtain an equation in the linear motion $x$.

### Solution

For the undamped energy method:

$$T+U=Constant\tag{1}$$

Where:

$T$ is the kinetic energy of the system

$U$ is the potential energy of the system

For our system, the total amount of kinetic energy is the energy from the rotating mass and the energy from the sliding mass:

$$T=\frac{1}{2}m\dot{x}^2+\frac{1}{2}J\dot{\theta}^2\tag{2}$$

An imporant relationship that we see here that we should remember is the arc length formula:

$$x=r\theta\tag{3}$$

This substitution rearranges equation 2 to read:

$$T=\frac{1}{2}\left(m+\frac{J}{r^2}\right)\dot{x}^2\tag{2a}$$

Next, we will look at the potential energy. In this case, we have two sources of potential energy, from the linear spring and from the torional stiffness off the steering rod:

$$U=\frac{1}{2}k_2x^2+\frac{1}{2}k_1\theta^2\tag{4}$$

Again, we can use equation 3 to rearange equation 4:

$$U=\frac{1}{2}\left(k_2+\frac{k_1}{r^2}\right)x^2\tag{4a}$$

Differentiating equation 1 gives us:

$$\frac{dT}{dt}+\frac{dU}{dt}=0\tag{1a}$$

Differentiating and plugging in equation 2a and 4a into equation 1a gives:

$$\left(m+\frac{J}{r^2}\right)\dot{x}\ddot{x}+\left(k_2+\frac{k_1}{r^2}\right)x\dot{x}=0\tag{1b}$$

This simplifies to:

$$M\ddot{x}+Kx=0\tag{1c}$$

Where:

Effective Mass: $M = m+\frac{J}{r^2}$

Effective Stiffness: $K=k_2+\frac{k_1}{r^2}$

## Problem 2

Consider the pendulum-and-spring system below:

Screen Shot 2021-09-21 at 10.11.37 PM.png

Here the mass of the pendulum rod is negligible. Derive the equation of motion using the energy method. Then linearize the system for small angles and determine the natural frequency. The length of the pendulum is $l$, the tip mass is $m$, and the spring system stiffness is $k$.

### Solution

We know that:

$$T+U=Constant\tag{1}$$

and in this equation:

$$T=\frac{1}{2}J\dot{\theta}^2\tag{2}$$

$$U=\frac{1}{2}k(lsin(\theta))^2+mg(lcos(\theta))\tag{3}$$

To isolate the equation in terms of mass and $\theta$ we will use the relation:

$$J=ml^2$$

Plugging this in gives:

$$T=\frac{1}{2}m(l\dot{\theta})^2\tag{2a}$$

Differentiating equation 1 and pluging in 2a and 3a gives:

$$k(lsin(\theta))(lcos(\theta))\dot{\theta}-mglsin(\theta)\dot\theta+ml\dot{\theta}(l\ddot{\theta})=0\tag{1a}$$

This equation simplifies to:

$$ml\ddot{\theta}-mgsin(\theta)+klsin(\theta)cos(\theta)=0\tag{1b}$$

We can linearize this equation with the first term of the taylor series expantion:

$$cos(\theta) = 1$$

$$sin(\theta) = \theta$$

Plugging these in gives an equation of motion:

$$ml\ddot{\theta}+(kl-mg)\theta=0$$

Because our equation follows a constant coefficent form, we can easily say the natural frequency $\omega_n=\sqrt{\frac{kl-mg}{ml}}$

## Problem 4

The displacement of a vibrating spring-mass-damper system is recorded on an $x-y$ plotter and reproduced in the figured below. The $y$ coordinate is the displacement in cm and the $x$ coordinate is time in seconds. From the plot determine the natural frequency, the damping ratio, and the damped natural frequency. 

Screen Shot 2021-09-22 at 7.50.47 PM.png

We know that with measured data:

$$\zeta = \frac{\delta}{\sqrt{4\pi^2-\delta^2}}$$

Where:

$$\delta=ln\left(\frac{y_1}{y_2}\right)$$

We also know that:

$$\omega_n = \frac{2\pi}{T}$$

Where $T$ is the period between 2 peaks directly next to eachother. 

```python
y_1 = 0.75e-2
y_2 = 0.21e-2

delta = np.log(y_1/y_2)

zeta = delta/np.sqrt(4*np.pi**2-delta**2)
print("zeta =", zeta)

zeta = zeta.round(2)

T = 4.875-0.875
omega_n = 2*np.pi/T
print("omega_n =", omega_n)

omega_d = omega_n*np.sqrt(1-zeta**2)
print("omega_d =",omega_d)
```
