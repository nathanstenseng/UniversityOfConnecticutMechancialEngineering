Problem 1:

Solve $m\ddot{x}+kx=0$ for $k=4$ N/m, $m=1$ kg, $x_o=1$ mm, and $v_o=0$ m/s. Plot the solution.

Solution:

$$m\ddot{x}+kx=0\tag{1a}$$
$$\ddot{x}+\frac{k}{m}x=0\tag{1b}$$
$$\omega^2 = \frac{k}{m}\tag{2}$$
$$\ddot{x}+\omega^2 x=0\tag{1c}$$

We know that equation 1c has a solution with the general form:

$$x(t)=A_1cos(\omega t)+A_2sin(\omega t)\tag{3}$$

We can solve for $A_1$ and $A_2$ using the given information and pluging them into equation 3 and its derivative.

$$x(0)=A_1cos(0)+A_2sin(0)$$
$$x'(0)=-\omega A_1sin(0)+\omega A_2cos(0)$$
$$ $$
$$\begin{bmatrix} 
cos(0) & sin(0) \\ 
-\omega sin(0) & \omega cos(0) 
\end{bmatrix}
\left[ \begin{array}{c} 
A_1 \\ 
A_2 
\end{array} \right]
 = 
 \left[ \begin{array}{c} 
 x(0) \\ 
 \dot{x}(0)
 \end{array} \right]$$

```python
import numpy as np
import matplotlib.pyplot as plt
x_0 = 0.001
v_0 = 0
k = 4
m = 1
omega = np.sqrt(k/m)

#AX = B     X = A^(-1)B
B = [[x_0],
     [v_0]]

A = [[np.cos(0),np.sin(0)],
     [-omega*np.sin(0),omega*np.cos(0)]]

X = np.linalg.inv(A).dot(B)

print("A_1 =",X[0])
print("A_2 =",X[1])

t = np.arange(0,10,0.1)
def get_x(t):
  return X[0]*np.cos(omega*t) + X[1]*np.sin(omega*t)

plt.plot(t, get_x(t))
plt.xlabel("Time (seconds)")
plt.ylabel("Distance (Meters)")
```

Problem 2:

An undamped system vibrates with a frequency of 10 Hz and amplitude 1 mm. Calculate the maximum amplitude of the system's velocity and acceleration.

Solution:

The equation of motion we will be looking at is:

$$x(t)=Asin(\omega_nt+\phi)$$
Where:

$A$ is amplitude

$\omega_n$ is the natural frequency

$\phi$ is the phase angle

$f=\frac{\omega}{2\pi}$

The maximum amplitude, based off the given information, would be 1 mm. Differentiating this equation will give us our velocity adn acceloration equation.

$$x(t)=sin(62.8t)\tag{1a}$$

$$\dot x(t) = 62.8cos(62.8t)\tag{1b}$$

$$\ddot x(t) = -3947sin(62.8t)\tag{1c}$$

Similarly, the velocity and acceloration maximum amplitudes are 62.8 mm/s and 3947 mm/s$^2$ respectively.
