## 1. Ideal Gas Law

On a very cold winter day, tom is getting ready to play football. In his cosy warm living room inflates a football to the NFL allowed value of 13 psi. The temperature in  the living room is 80$^o$F. Before going out to meet the rest of the team, he checks the weather app on his phone and finds out that the temperature outside is 16$^o$F and the atmospheric pressure is 29.5 inHg (inches of mercury). What is the pressure in Tomâ€™s football in psi when he is outside? Assume that the volume of the football does not change.

### Solution

The main equation dictating the pressure of the football here is the Ideal Gas Law:

$$pv=RT\tag{1a}$$

We know that the volume is constant so we will set the equation equal to $v$:

$$v=\frac{RT_{1,abs}}{p_{1,abs}}\tag{1b}$$

We are given that $T_1=80^oF$ and $p_1 = 13psi$. We need to change these to absolute terms though. 

$$p_{1}=p_{1,abs}-p_{ambient}\tag{2}$$

$$T_{1,abs} = T_1 + 459.67\tag{3}$$

Now we have values of $T_{1,abs} = 539.67^oR$. To solve equaiton 2, we will need to use $1 psi = 144\frac{lb_f}{ft^2}$ and $1 inHg = 70.73 \frac{lb_f}{ft^2}$. Using these values to solve equation 2 gives $p_{1,abs}=3958.5\frac{lb_f}{ft^2}$. Now we can use our Ideal Gas Law equation:

$$v=\frac{53.353\frac{ft\cdot lb_f}{lb\cdot  ^oR}\cdot 539.67^oR}{3958.5\frac{lb_f}{ft^2}}\tag{1b}=7.27\frac{ft^3}{lb}$$

We can now rearrange the ideal gas law to better suit the second half of the equation:

$$p_{2,abs} = \frac{RT_{2,abs}}{v}\tag{1c}$$

We will again use equation 3 to get $T_{2,abs} = 475.67$. Using equation 1C now gives

$$p_{2,abs} = \frac{53.353\frac{ft\cdot lb_f}{lb\cdot  ^oR}\cdot 475.67^oR}{7.27\frac{ft^3}{lb}} = 3490.84 \frac{lb_f}{ft^2}$$

Using equation 2, we get that $p_2 = 1404 \frac{lb_f}{ft^2}=9.74 psi$.

## 2. Stress and Force

The goal of this problem is to show that stress in the fluid varies in space (in this case with respect to the x coordinate). Also, to demonstrate that stress causes a force.

Air flows through two vertical plates as shown in Firugre 2. The velocity of air between in unites of m/s depends only on the x-coordinate and it is given by

$$u(x)=\left(\frac{1}{4}-x^2\right)\tag{1}$$

where the left plate is at $x=-\frac{1}{2}$ m and the right at $x=-\frac{1}{2}$ m, as in the figure below.

(a) Find the shear stress as a function of $x$ if the dynamic viscosity is $\mu =1.4\times10^{-5}$ Pa$\cdot$s. We will discuss late in the class that there is only one component of shear stress, which is on the $x$-$z$ plane and it is given by $\tau=\mu\frac{\partial u}{\partial x}$

(b) Find the magnitude of the force exerted by the fluid on each of the two plates, if the plates have dimension $L_y=0.25$ m and $L_z=0.5$ m.

(c) What is the direction of the force on each of the plates?

Screen Shot 2021-09-13 at 3.30.46 PM.png

#### Solution

(a) To find shear stress, we will use 

$$\tau =\mu\frac{\partial u}{\partial x}$$

Since $u$ is already a function of $x$ purley, our partial derivative becomes an ordinary derivative:

$$\frac{du}{dx}=-2x$$

Plugging in this and given values gives:

$$\tau(x)=-2.8x\times10^{-5}$$

This equation will have the units of Pa since $\mu$ has units Pa$\cdot$s and $\frac{du}{dx}$ has units of $\frac{m/s}{m}$.

(b) To find the stress on both plates, we will plug in the coordinate location to our stress equation:

$$\tau\left(-\frac{1}{2}\right)=1.4\times10^{-5} Pa$$

$$\tau\left(\frac{1}{2}\right)=-1.4\times10^{-5} Pa$$

To find the force on each plate, we will use:

$$F = \tau A$$

Using this, with an area of $0.125$ m$^2$, we find that the magnitude of the force on both plates will be $0.175\times10^{-6}$ N.

(c) The direction of the force on both plates will be in the positive z direction. This is because on the right plate, the negative stress means the internal fluid forces will be moving down the wall, and the wall will have an opposite reaction. This applies similary for the left wall.

## 3. Viscosity

True story: a couple of years ago an ME Senior Design Project team was working with a company that manufactures fuel pumps to construct a viscometer. The viscometer is an instrument that measures a fluidâ€™s viscosity. The solution the team found was to measure the force on a section of a cylindrical tube. From the force measurement they would calculate the viscosity. Also, the team was able to find the velocity distribution which is:

$$u(r)=0.01\left[1-\left(\frac{r}{0.005}\right)^2\right]m/s\tag{2}$$

The fluid velocity is expressed in cylindrical coordinates as in Figure 3. The tube radius is $R = 0.005$ m and the tube length is $l = 1$ m. For the configuration of the flow in the circular pile, there is only one component of the shear stress on the xâ€“r plane and it is given by (in cylindrical coordinates):

$$\tau(r)=\mu\frac{\partial u}{\partial r}\tag{3}$$

If the force on the pipe is $F=10^{-3}\pi$ N, what is the value of the dynamic viscosity $\mu$?

### Solution

First, to convert our force to make it compatible with stress, we will use the fact the the surface area of our open pipe is:

$$A=2\pi r l$$

Now we can use our stress equation to say:

$$\tau = \frac{-10^{-3}\pi}{2\pi(0.005)(1)}=-0.1$$

Plugging this value of $\tau$ into equation 3 can let us find $\mu$, but first we will need to calculate:

$$\frac{\partial u}{\partial r}=\frac{du}{dr}=0.01\left[-2\left(\frac{r}{0.005}\right)\left(\frac{1}{0.005}\right)\right]=-4\frac{1}{s}$$

Now we can rearrange equation 3 to say:

$$\mu=\frac{\tau}{\frac{\partial u}{\partial r}}=\frac{-0.1}{-4}=0.025Pa\cdot s$$
